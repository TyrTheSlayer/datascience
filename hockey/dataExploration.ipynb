```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
skaters = pd.read_csv("~/datascience/hockey/skaters.csv")
skaters.head()
```

The data is organized as so:
**Player**: Player name
**Pos**: Position between C (center), LW (Left winger), RW (Right winger), and D (defenseman)
**Tm**: Team the player played for
**Season**: Season the data line was from (October of first year to ~April of next year)
**GP**: Games played
**G**: Goals
**A**: Assists
**PTS**: Points

Corsi stat (Shots + Blocks + Misses):
**CF**: Corsi For the player 
**CA** Corsi against the player
**CF%**: CF / (CF + CA). Above 50% means the team was controlling the puck more often than not with this player on the ice
**CF% rel**: CF on ice% - CF off ice %
**C/60**: Corsi per 60 minutes in all situations (CF - CA) * 60 / Time on Ice
**Crel/60**: CF/60 - CFoff ice /60

Fenwick stat (Shots + misses):
**FF**: Fenwick for the player
**FA**: Fenwick against the player
**FF%**: FF/(FF + FA). 
**FF% rel**: FF on ice% - FF off ice %

PDO:
**oiSH%**: Team on ice shooting percentage while this player is on the ice
**oiSV%**: Team on ice save percentage while this player is on the ice
**PDO**: Shooting% + Save% 

Zone Starts:
**oZS%**: Offensive zone start & in all situations while this player is on the ice. Offensive zone Faceoffs / (Offensive zone Faceoffs + Defensive zone Faceoffs)
**dZS%**: Defensive zone start & in all situations while this player is on the ice. Defensive zone Faceoffs / (Offensive zone Faceoffs + Defensive zone Faceoffs)
**TOI/Gm**: Average time on ice per game
**FOW**: Face off wins
**FOL**: Face off losses
**FO%**: Face off win percentage
**HIT**: Number of hits
**BLK**: Number of blocks
**TK**: Number of takeaways
**GV**: Number of Giveaways

Trophies (1 if earned in that season, 0 if not):
**Hart**: Hart Memorial trophy, awarded to the "player judged most valuable to his team". 
**Byng**: Lady Byng Memorial Trophy, awarded to the player who exhibited outstanding sportsmanship and gentlemanly conduct combined with a high standard of playing ability.
**Norris**: James Norris Memorial Trophy, awarded to the defenseman who demonstrates throughout the season the greatest all-round ability in the position.
**Lindsay**: Ted Lindsay Award, awarded to the NHL's outstanding player as selected by the members of the NHL Players Association
**Selke**: Frank J. Selke Trophy, awarded to the forward who best excels in the defensive aspects of the game.
